## Is it a bird?

```python
from fastbook import *
path = Path('bird_or_not')
```

```python
urls = search_images_ddg('bird photos', max_images=1)
len(urls),urls[0]
```

```python
dest = 'bird.jpg'
download_url(urls[0], dest, show_progress=False)
```

```python
im = Image.open(dest)
im.to_thumb(256,256)
```

All models need numbers as their inputs. Images are already stored as numbers in a computer -- here's a viewer so you can see what that looks like: [PixSpy](https://pixspy.com/).

```python
searches = 'forest','bird'

for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True)
    results = search_images_ddg(f'{o} photo')
    download_images(dest, urls=results[:200])
```

```python
failed = verify_images(get_image_files(path))
failed.map(Path.unlink);
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path)

dls.show_batch(max_n=6)
```

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

```python
is_bird,_,probs = learn.predict(PILImage.create('bird.jpg'))
print(f"This is a: {is_bird}.")
print(f"Probability it's a bird: {probs[0]:.4f}")
```

## Overfitting

```python
plt.rc('figure', dpi=90)

def plot_function(f, title=None, min=-2.1, max=2.1, color='r', ylim=None):
    x = torch.linspace(min,max, 100)[:,None]
    if ylim: plt.ylim(ylim)
    plt.plot(x, f(x), color)
    if title is not None: plt.title(title)
```

```python
def f(x): return 3*x**2 + 2*x + 1

plot_function(f, "$3x^2 + 2x + 1$")
```

```python
from numpy.random import normal,seed,uniform
np.random.seed(42)

def noise(x, scale): return normal(scale=scale, size=x.shape)
def add_noise(x, mult, add): return x * (1+noise(x,mult)) + noise(x,add)
```

```python
x = torch.linspace(-2, 2, steps=20)[:,None]
y = add_noise(f(x), 0.3, 1.5)
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def plot_poly(degree):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(x, y)
    plt.scatter(x,y)
    plot_function(model.predict)
```

## How do we fit a function to data?

```python
def quad(a, b, c, x): return a*x**2 + b*x + c
def mk_quad(a,b,c): return partial(quad, a,b,c)
```

```python
f = mk_quad(2,3,4)
quad(2,3,4,1.5), f(1.5)
```

```python
from ipywidgets import interact
```

```python
@interact(a=1.5, b=1.5, c=1.5)
def plot_quad(a, b, c):
    plt.scatter(x,y)
    plot_function(mk_quad(a,b,c), ylim=(-3,12))
```

```python
def mse(preds, acts): return ((preds-acts)**2).mean()
```

```python
@interact(a=1.5, b=1.5, c=1.5)
def plot_quad(a, b, c):
    f = mk_quad(a,b,c)
    plt.scatter(x,y)
    loss = mse(f(x), y)
    plot_function(f, ylim=(-3,12), title=f"MSE: {loss:.2f}")
```

```python
def quad_mse(params):
    f = mk_quad(*params)
    return mse(f(x), y)
```

```python
quad_mse([1.5, 1.5, 1.5])
```

```python
abc = torch.tensor([1.5,1.5,1.5])
abc.requires_grad_()
```

```python
loss = quad_mse(abc)
loss
```

```python
loss.backward()
```

```python
abc.grad
```

```python
with torch.no_grad():
    abc -= abc.grad*0.01
    loss = quad_mse(abc)
    
print(f'loss={loss:.2f}')
```

```python
for i in range(5):
    loss = quad_mse(abc)
    loss.backward()
    with torch.no_grad(): abc -= abc.grad*0.01
    print(f'step={i}; loss={loss:.2f}')
```

```python
def rectified_linear(m,b,x):
    y = m*x+b
    return torch.clip(y, 0.)
```

```python
plot_function(partial(rectified_linear, 1,1))
```

```python
@interact(m=1.5, b=1.5)
def plot_relu(m, b):
    plot_function(partial(rectified_linear, m,b), ylim=(-1,4))
```

```python
def double_relu(m1,b1,m2,b2,x):
    return rectified_linear(m1,b1,x) + rectified_linear(m2,b2,x)
```

```python
@interact(m1=-1.5, b1=-1.5, m2=1.5, b2=1.5)
def plot_double_relu(m1, b1, m2, b2):
    plot_function(partial(double_relu, m1,b1,m2,b2), ylim=(-1,6))
```
